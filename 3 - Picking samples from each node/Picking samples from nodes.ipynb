```python
import pandas as pd
import numpy as np
import os
```

```python
# function generates equal ranged index list for df shape 
# for picking sample data from df

def get_ranged_index_list(df_size_, sample_size_):
    if(df_size_ <= sample_size_):
        return df.index.tolist()
    else:
        c = 0
        range_ = int(df_size_/sample_size_)

        index_list_ = []
        for i in range(sample_size_):
            index_list_.append(c)
            c+=range_

        return index_list_
```

```python
# Read csv files (chunked descriptionary) and pick sample data from each

#path = '/Users/altay.amanbay/Desktop/new node booster/experiments/3/1 - Descriptionary nodes into separate csv files/Descriptionary nodes in separate csv files/'
path = '/Users/altay.amanbay/Desktop/new node booster/experiments/train data from descriptionary nodes by sampling/1 - Descriptionary nodes into separate csv files/Descriptionary nodes in separate csv files/'
print('Reading from following path:')
print(path,'\n')

samples_df = None
sample_size = 5000

# iterate through csv files
for file in os.listdir(path):
    # Pick only csv files
    if file.endswith(".csv"):
        if(file == '1-332.csv'):
            continue
        
        # Read data from csv file
        print('Reading file %s ...' % file)
        df = pd.read_csv(os.path.join(path, file))
        print('Data shape ', df.shape[0])
        #print(os.path.join(path, file))

        # Sort data by description
        print('Sorting file %s by description  ...' % file)
        df.sort_values(by=['description'], ascending=True, inplace=True)
        df.reset_index(drop=True, inplace=True)

        # Pick sample data
        print('Picking sample ...')
        index_list = get_ranged_index_list(df.shape[0], sample_size)
        df = df.loc[index_list,['description','category_id','category_path']]

        # Concatenate samples
        if(samples_df is None):
            samples_df = df.copy(deep=True)
            print('shape',samples_df.shape)
        else:
            samples_df = pd.concat([samples_df,df],axis=0)
            print('shape',samples_df.shape)
            
        print()

        
# Convert index into column
samples_df.reset_index(drop=False, inplace=True)
        
# Export sampled df into csv
path = '/Users/altay.amanbay/Desktop/new node booster/experiments/train data from descriptionary nodes by sampling/3 - Picking samples from each node/sampled descriptionary/'
file_name = 'sampled_descriptionary_sample_size_'+str(sample_size)+'.csv'
samples_df.to_csv(path + file_name, index=False)
```
